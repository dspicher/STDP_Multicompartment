```python
%pylab inline
import cPickle
from helper import PeriodicAccumulator, BooleanAccumulator
from itertools import product
```

```python
def get(weight):
    return cPickle.load(open('characterize_epsp_weight_{0}.p'.format(weight),'rb'))


weights = [1e-05, 2e-05, 5e-05, 0.0001, 0.0002, 0.0005, 0.001, 0.002, 0.005, 0.01, 0.02, 0.05, 0.1, 0.2, 0.5]


params = list(product(weights))

data = {tup:get(*tup) for tup in params}
```

```python
qtts = ['somatic','dendritic','prediction']
figure(figsize=(12,5))
for idx, qtt in enumerate(qtts):
    maxs = []
    for p in params:
        maxs.append(np.max(data[p][0].res['y'][:,idx]))
    semilogx([p[0] for p in params], maxs)
legend(qtts,loc=0)
xlabel('weight')
title('peak value')

figure(figsize=(12,5))
for idx, qtt in enumerate(qtts):
    maxs = []
    for p in params:
        maxs.append(-30+data[p][0].t[np.argmax(data[p][0].res['y'][:,idx])])
    semilogx([p[0] for p in params], maxs)
legend(qtts,loc=0)
xlabel('weight')
title('peak time after epsp start')
```
