```python
   
Us = np.arange(-70,-20,0.01)
def plot_exp(U0,a,shift):
    semilogy(Us,1.0/(U0*np.exp(a*(Us-shift))))
    xlim([-70,-20])
    


i = interact(plot_exp,
         U0=(0.001,1,0.001),
         a=(0.,4.0,0.001),
         shift=(-70,-20)
         )

```
