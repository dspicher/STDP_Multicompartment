```python
%pylab inline
import cPickle
from helper import PeriodicAccumulator, BooleanAccumulator
from itertools import product
```

```python
def get(alpha, beta, r_max, thresh, tau_ref_0, theta_0, Uclamp):
    return cPickle.load(open('artola_wiggle_alpha_{0}_beta_{1}_r_max_{2}_thresh_{3}_tau_ref_0_{4}_theta_0_{5}_Uclamp_{6}.p'.format(alpha, beta, r_max, thresh, tau_ref_0, theta_0, Uclamp),'rb'))


alpha_s = [-32.3, -34.0, -35.7]
beta_s = [0.2375, 0.25, 0.2625]
r_max_s = [0.02375, 0.025, 0.026250000000000002]
thresh_s = [-19.0, -20.0, -21.0]
tau_ref_0_s = [4.75, 5.0, 5.25]
theta_0_s = [2.375, 2.5, 2.625]
Uclamp_s = array([-40., -30., -20., -10.,   0.])


params = list(product(alpha_s, beta_s, r_max_s, thresh_s, tau_ref_0_s, theta_0_s, Uclamp_s))

data = {tup:get(*tup) for tup in params}
```

```python
artola_data = np.loadtxt(open('./experimental_data/artola_voltage_clamp.csv','r'),delimiter=',')
artola_data[:,1] = artola_data[:,1] - 1
artola_data[:,1] = artola_data[:,1]/artola_data[-1,1]
```

```python
rel_res = {}
params2 = list(product(alpha_s, beta_s, r_max_s, thresh_s, tau_ref_0_s, theta_0_s))
for p_idx,partial_p in enumerate(params2):
    dws = np.zeros(len(Uclamp_s))
    for idx,U2 in enumerate(Uclamp_s):
        p = tuple(list(partial_p)+[U2])
        dws[idx] = data[p][0].res['weight'][-1]/data[p][0].res['weight'][0]
    dws = dws - 1
    dws = dws/dws[-1]
    rel_res[partial_p] = dws
```

```python
errors = {k:np.sum(np.abs(rel_res[k]-artola_data[:,1])) for k in rel_res.keys()}
```

```python
ks,errs = zip(*errors.items())
sort_idxs = argsort(errs)
for i in range(20):
    figure(figsize=(12,5))
    dws = rel_res[ks[sort_idxs[i]]]
    plot(Uclamp_s, dws,'.--')
    scatter(Uclamp_s, dws, lw=0, c='red', s=50)
    xlim([-45,5])
    scatter(artola_data[:,0],artola_data[:,1],c='g',s=50)
    title("{0}".format(ks[sort_idxs[i]]))
    
```

```python
def show_plot(key,alpha, beta, r_max, thresh, tau_ref_0, theta_0, Uclamp,y_c,t_min,t_max):
    figure(figsize=(12,5))
    p = (alpha, beta, r_max, thresh, tau_ref_0, theta_0, Uclamp)
    ts = data[p][0].t
    mask = np.logical_and(ts>=t_min,ts<=t_max)
    if key=='y':
        plot(data[p][0].t[mask],data[p][0].res[key][mask,y_c])
    else:
        plot(data[p][0].t[mask],data[p][0].res[key][mask])
```
