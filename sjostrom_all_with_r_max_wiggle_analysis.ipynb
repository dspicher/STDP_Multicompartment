```python
%pylab inline
import cPickle
from helper import PeriodicAccumulator, BooleanAccumulator
from itertools import product
```

```python
def get(alpha, beta, g_L):
    return cPickle.load(open('sjostrom_curves_alpha_{0}_beta_{1}_g_L_{2}.p'.format(alpha, beta, g_L),'rb'))


alpha_s = array([-60., -58., -56., -54., -52., -50.])
beta_s = [0.1, 0.2]
g_L_s = [0.05, 0.03]


params = list(product(alpha_s, beta_s, g_L_s))

data = {tup:get(*tup) for tup in params}
```

```python
for p in params:
    if p[1]==0.2:
        continue
    if p[2]==0.05:
        continue
    if len(data[p])==0:
        continue
    for r_idx,arr in enumerate(data[p][0]):
        figure(figsize=(12,5))
        for idx in range(arr.shape[2]):
            plot([1,10,20,40,50],arr[:,0,idx],'b')
            plot([1,10,20,40,50],arr[:,1,idx],'r')
        title(str(p)+" r"+str(r_idx))
```
